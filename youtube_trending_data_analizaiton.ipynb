ê°œë³„ ì› ë°ì´í„° ì‚¬ìš© ì‹œ ê°’ì´ ì ë¦° ê³³ì´ ë§ì´ ì¡´ì¬ -> ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ì§€ ì•ŠìŒ
ì¡°íšŒìˆ˜, ì¢‹ì•„ìš”ìˆ˜, ì‹«ì–´ìš”ìˆ˜ ë³€ê²½ê°’ì˜ í‰ê· ìœ¼ë¡œ ì¸ê¸‰ë™ ì¬ì§„ì… í™•ì¸

-> ì‹œê°„ì´ ì§€ë‚ ìˆ˜ë¡ ê´€ì‹¬ì´ ì§€ì†ëëŠ”ì§€ë¥¼ í™•ì¸í•˜ê¸° ìœ„í•´ í‰ê·  ì‚¬ìš©

```python
# ë¹„ìœ¨ êµ¬í•˜ê¸° ìœ„í•œ ë³€ìˆ˜ 1
# ë³€í™”ëŸ‰ ë¨¼ì € ê³„ì‚°
df['view_diff'] = df.groupby('video_id')['view_count'].diff()
df['likes_diff'] = df.groupby('video_id')['likes'].diff()
df['dislikes_diff'] = df.groupby('video_id')['dislikes'].diff()
df['comment_count_diff'] = df.groupby('video_id')['comment_count'].diff()

# video_idë³„ diff í‰ê·  ê³„ì‚° í›„ ê° í–‰ì— ë¶™ì´ê¸°
df['view_count_diff_mean'] = df.groupby('video_id')['view_diff'].transform(lambda x: x.dropna().mean())
df['likes_diff_mean'] = df.groupby('video_id')['likes_diff'].transform(lambda x: x.dropna().mean())
df['dislikes_diff_mean'] = df.groupby('video_id')['dislikes_diff'].transform(lambda x: x.dropna().mean())
df['comment_count_diff_mean'] = df.groupby('video_id')['comment_count_diff'].transform(lambda x: x.dropna().mean())
```

```python
# ë¹„ìœ¨ êµ¬í•˜ê¸° ìœ„í•œ ë³€ìˆ˜ 2
# groupby + aggë¡œ ì²˜ìŒê³¼ ë§ˆì§€ë§‰ ê°’ ì°¨ì´ ê³„ì‚°
agg_df = df.groupby('video_id').agg({
    'view_count': lambda x: x.iloc[-1] - x.iloc[0],
    'likes': lambda x: x.iloc[-1] - x.iloc[0],
    'dislikes': lambda x: x.iloc[-1] - x.iloc[0],
    'comment_count': lambda x: x.iloc[-1] - x.iloc[0]
}).rename(columns={
    'view_count': 'view_count_total_diff',
    'likes': 'likes_total_diff',
    'dislikes': 'dislikes_total_diff',
    'comment_count': 'comment_count_total_diff'
})

# ì›ë³¸ dfì— ë³‘í•©
df = df.merge(agg_df, on='video_id', how='left')
```

```python
# ì™œë„ì™€ ì²¨ë„ ê³„ì‚° í•¨ìˆ˜
def calculate_skew_kurtosis(df):
    result = []

    for column in df.select_dtypes(include=[np.number]).columns:
        skew_val = df[column].skew()
        kurt_val = df[column].kurtosis()
        result.append({
            'feature': column,
            'skewness': round(skew_val, 3),
            'kurtosis': round(kurt_val, 3)
        })

    return pd.DataFrame(result).sort_values(by=['feature'], ascending=False)


# ì™œë„ì™€ ì²¨ë„ ê°’ ì¶œë ¥
calculate_skew_kurtosis(df)

# ë‹¤ì–‘í•œ ë³€í™˜ ì ìš© í•¨ìˆ˜
def transform_features(df):
    for column in cols:
        if column == 'video_id' or column == 'trending_count':
            continue
        
        skewness = df[column].skew()
        kurt = df[column].kurtosis()
        
        # ì ˆëŒ€ ì™œë„ì™€ ì²¨ë„ê°€ íŠ¹ì • ì„ê³„ê°’ì„ ë„˜ëŠ” ê²½ìš° ë³€í™˜ ì‹œë„
        if abs(skewness) > 3 or abs(kurt) > 10 or abs(skewness) < -3 or abs(kurt) < -10 :
            # print(f'Transforming {column} (skewness={skewness}, kurtosis={kurt})')
            # ë¡œê·¸ ë³€í™˜
            df.loc[:, column + '_log'] = np.log1p(df[column] - df[column].min() + 1)
            # ì œê³±ê·¼ ë³€í™˜
            df.loc[:, column + '_sqrt'] = np.sqrt(df[column] - df[column].min() + 1)
            # ì—­ìˆ˜ ë³€í™˜
            df.loc[:, column + '_reciprocal'] = 1 / (df[column] - df[column].min() + 1)

    return df


cols = [
    # ID
    'video_id',
    # Target
    'trending_count',
    
    'title_length',
    'hours_to_trend',
    'tag_count',
    'view_count_ratio',
    'likes_ratio',
    'dislikes_ratio',
    'comment_count_ratio',
    'description_length'
]
# ë³€í™˜ ì ìš©
df_skew_kurt = transform_features(df[cols])

# ë³€í™˜ í›„ ì™œë„ì™€ ì²¨ë„ ê°’ ì¶œë ¥
df_skew_kurt_printer = calculate_skew_kurtosis(df_skew_kurt)
df_skew_kurt_printer
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



def label_encoding(df, columns):
    df_copy = df.copy()
    class_maps = {}

    for column in columns:
        if pd.api.types.is_integer_dtype(df_copy[column]):
            df_copy[column] = df_copy[column].astype('category')
        else:
            le = LabelEncoder()
            encoded = le.fit_transform(df_copy[column])
            df_copy[column] = pd.Series(encoded, index=df_copy.index).astype('category')
            class_maps[column] = le.classes_

    return df_copy, class_maps




def one_hot_encoding(df, columns, drop_first=False):
    # í•­ìƒ ë³µì‚¬ë³¸ìœ¼ë¡œ ì‘ì—…
    df_copy = df.copy()

    # get_dummies() ì‹¤í–‰
    df_copy = pd.get_dummies(df_copy, columns=columns, drop_first=drop_first)

    # True/False â†’ 1/0 ë³€í™˜ (bool â†’ int)
    bool_cols = df_copy.select_dtypes(include=['bool']).columns
    df_copy[bool_cols] = df_copy[bool_cols].astype(int)

    return df_copy
```

## íšŒê·€ ëª¨ë¸ ì§€í‘œ í•´ì„ ê°€ì´ë“œ
- **MAE (Mean Absolute Error)**  
  - í‰ê· ì ìœ¼ë¡œ ì˜ˆì¸¡ê°’ì´ ì‹¤ì œê°’ê³¼ ì–¼ë§ˆë‚˜ ì°¨ì´ ë‚˜ëŠ”ì§€  
  - ì‘ì„ìˆ˜ë¡ ì¢‹ìŒ

- **MAPE (Mean Absolute Percentage Error)**  
  - ì‹¤ì œê°’ ëŒ€ë¹„ ì˜ˆì¸¡ ì˜¤ì°¨ì˜ í‰ê·  ë¹„ìœ¨ (%)  
  - 0~10%: ë§¤ìš° ìš°ìˆ˜ / 10~20%: ì–‘í˜¸ / 20~50%: ë³´í†µ / 50% ì´ìƒ: ë‚˜ì¨

- **RMSE (Root Mean Squared Error)**  
  - í° ì˜¤ì°¨ì— ë” ë¯¼ê°í•œ í‰ê·  ì˜¤ì°¨  
  - ì‘ì„ìˆ˜ë¡ ì¢‹ìŒ  
  - MAEë³´ë‹¤ í•­ìƒ í¬ê±°ë‚˜ ê°™ìŒ

- **RÂ² (R-squared)**  
  - 0~1 ì‚¬ì´ ê°’ (1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ëª¨ë¸ì´ ë°ì´í„°ë¥¼ ì˜ ì„¤ëª…í•¨)  
  - 0.7 ì´ìƒì´ë©´ ëŒ€ì²´ë¡œ ì„¤ëª…ë ¥ ìš°ìˆ˜

- **Adjusted RÂ²**  
  - ì„¤ëª…ë³€ìˆ˜ê°€ ë§ì„ìˆ˜ë¡ RÂ²ê°€ ê³¼í•˜ê²Œ ë†’ì•„ì§€ëŠ” ê±¸ ë³´ì •  
  - RÂ²ë³´ë‹¤ ì¡°ê¸ˆ ë‚®ê±°ë‚˜ ë¹„ìŠ·í•œ ìˆ˜ì¤€ì´ì–´ì•¼ ì •ìƒ  
  - **â†’ ë¹„ì •ìƒì ìœ¼ë¡œ ë†’ê±°ë‚˜ ë‚®ìœ¼ë©´ ê³„ì‚° ì˜¤ë¥˜ ê°€ëŠ¥**

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error


def get_mae(model, X_test, y_test):
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    print('{0} MAE: {1}'.format(model.__class__.__name__, np.round(mae, 3)))
    return mae

def get_mape(model, X_test, y_test):
    pred = model.predict(X_test)
    # ì˜ˆì¸¡ê°’ì´ 0ì— ê°€ê¹Œìš°ë©´ ë¬¸ì œê°€ ë˜ë¯€ë¡œ ì‘ì€ ìˆ˜ ë”í•¨
    mape = np.mean(np.abs((y_test - pred) / (y_test + 1e-8))) * 100
    print('{0} MAPE: {1:.2f}%'.format(model.__class__.__name__, mape))
    return mape

def get_rmse(model, X_test, y_test):
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test , pred)
    rmse = np.sqrt(mse)
    print('{0} ë¡œê·¸ ë³€í™˜ëœ RMSE: {1}'.format(model.__class__.__name__,np.round(rmse, 3)))
    return rmse

def get_r2(model, X_test, y_test):
    pred = model.predict(X_test)
    r2 = r2_score(y_test , pred)
    print('{0} RÂ²: {1}'.format(model.__class__.__name__,np.round(r2, 3)))
    return r2

def get_adjusted_r2(model, X_test, y_test, n_features):
    pred = model.predict(X_test)
    r2 = r2_score(y_test, pred)
    n = len(y_test)
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - n_features - 1)
    print('{0} Adjusted RÂ²: {1:.3f}'.format(model.__class__.__name__, adj_r2))
    return adj_r2

def get_evaluation_metrics(model, X_test, y_test, n_features):
    mae = get_mae(model, X_test, y_test)
    mape = get_mape(model, X_test, y_test)
    rmse = get_rmse(model, X_test, y_test)
    r2 = get_r2(model, X_test, y_test)
    adj_r2 = get_adjusted_r2(model, X_test, y_test, n_features)
    return mae, mape, rmse, r2, adj_r2
```

  - 0.6 ì´ìƒ: ë³´í†µ / 0.7 ì´ìƒ: ì–‘í˜¸ / 0.8 ì´ìƒ: ìš°ìˆ˜

- **ROC AUC (Area Under Curve)**  
  - ëª¨ë¸ì˜ í´ë˜ìŠ¤ êµ¬ë¶„ ëŠ¥ë ¥ì„ ë‚˜íƒ€ëƒ„  
  - 0.5 = ë¬´ì‘ìœ„ / 1.0 = ì™„ë²½í•œ ë¶„ë¦¬  
  - 0.7~0.8: ë³´í†µ / 0.8~0.9: ì–‘í˜¸ / 0.9 ì´ìƒ: ë§¤ìš° ìš°ìˆ˜  
  - ë‹¤ì¤‘ ë¶„ë¥˜ì—ì„œëŠ” OVO ë°©ì‹ ë“±ìœ¼ë¡œ í™•ì¥ ê°€ëŠ¥


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
import numpy as np

def get_accuracy(model, X_test, y_test):
    pred = model.predict(X_test)
    acc = accuracy_score(y_test, pred)
    print('{0} Accuracy: {1:.3f}'.format(model.__class__.__name__, acc))
    return acc

def get_precision(model, X_test, y_test):
    pred = model.predict(X_test)
    prec = precision_score(y_test, pred, average='macro', zero_division=0)
    print('{0} Precision (macro): {1:.3f}'.format(model.__class__.__name__, prec))
    return prec

def get_recall(model, X_test, y_test):
    pred = model.predict(X_test)
    rec = recall_score(y_test, pred, average='macro', zero_division=0)
    print('{0} Recall (macro): {1:.3f}'.format(model.__class__.__name__, rec))
    return rec

def get_f1(model, X_test, y_test):
    pred = model.predict(X_test)
    f1 = f1_score(y_test, pred, average='macro', zero_division=0)
    print('{0} F1 Score (macro): {1:.3f}'.format(model.__class__.__name__, f1))
    return f1

def get_roc_auc(model, X_test, y_test):
    try:
        proba = model.predict_proba(X_test)
        if proba.shape[1] == 2:
            auc = roc_auc_score(y_test, proba[:, 1])
        else:
            auc = roc_auc_score(y_test, proba, multi_class='ovo', average='macro')
        print('{0} ROC AUC: {1:.3f}'.format(model.__class__.__name__, auc))
        return auc
    except:
        print('{0} ROC AUC: Not Available (no predict_proba)'.format(model.__class__.__name__))
        return None

def get_classification_metrics(model, X_test, y_test):
    acc = get_accuracy(model, X_test, y_test)
    prec = get_precision(model, X_test, y_test)
    rec = get_recall(model, X_test, y_test)
    f1 = get_f1(model, X_test, y_test)
    auc = get_roc_auc(model, X_test, y_test)
    return acc, prec, rec, f1, auc
```

```python
def print_classification_metrics(acc, prec, rec, f1, auc):
    print("ë¶„ë¥˜ ëª¨ë¸ í‰ê°€ ê²°ê³¼")
    print("="*40)
    print(f"Accuracy (ì •í™•ë„): {acc:.3f}")
    print("    â†’ ì „ì²´ ì˜ˆì¸¡ ì¤‘ ì •ë‹µ ë¹„ìœ¨")

    print(f"Precision (ì •ë°€ë„, macro): {prec:.3f}")
    print("    â†’ ì˜ˆì¸¡í•œ ê²ƒ ì¤‘ ì‹¤ì œë¡œ ë§ì€ ë¹„ìœ¨ (í´ë˜ìŠ¤ë³„ í‰ê· )")

    print(f"Recall (ì¬í˜„ìœ¨, macro): {rec:.3f}")
    print("    â†’ ì‹¤ì œ ì •ë‹µ ì¤‘ ì˜ˆì¸¡ì´ ë§ì€ ë¹„ìœ¨ (í´ë˜ìŠ¤ë³„ í‰ê· )")

    print(f"F1 Score (macro): {f1:.3f}")
    print("    â†’ ì •ë°€ë„ì™€ ì¬í˜„ìœ¨ì˜ ì¡°í™” í‰ê· ")

    if auc is not None:
        print(f"ROC AUC: {auc:.3f}")
        print("    â†’ í´ë˜ìŠ¤ ë¶„ë¦¬ ëŠ¥ë ¥ (1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ìš°ìˆ˜)")
    else:
        print("ROC AUC: ì œê³µ ë¶ˆê°€ (predict_proba ì—†ìŒ)")

    print("="*40)
```

```python
def plot_logistic_coefficients(model, X, class_labels=None, figsize=(10, 6), top_n=20):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    if not hasattr(model, "coef_"):
        raise ValueError("í•´ë‹¹ ëª¨ë¸ì€ coef_ ì†ì„±ì´ ì—†ìŠµë‹ˆë‹¤.")
    
    coef = model.coef_  # shape: (n_classes, n_features)
    n_classes = coef.shape[0]
    features = X.columns

    class_labels = class_labels or [f"class {i}" for i in range(n_classes)]

    for i in range(n_classes):
        coef_df = pd.DataFrame({
            'feature': features,
            'coefficient': coef[i]
        })
        top_coef = coef_df.reindex(
            coef_df['coefficient'].abs().sort_values(ascending=False).index
        ).head(top_n)

        plt.figure(figsize=figsize)
        plt.barh(top_coef['feature'], top_coef['coefficient'], 
                 color=np.where(top_coef['coefficient'] > 0, 'skyblue', 'salmon'))
        plt.xlabel('ê³„ìˆ˜ê°’')
        plt.title(f'Logistic Regression Coefficients for {class_labels[i]}')
        plt.gca().invert_yaxis()
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
def plot_logistic_odds_ratios(model, X, class_labels=None, figsize=(10, 6), top_n=20, exclude_features=None):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    if not hasattr(model, "coef_"):
        raise ValueError("í•´ë‹¹ ëª¨ë¸ì€ coef_ ì†ì„±ì´ ì—†ìŠµë‹ˆë‹¤.")
    
    coef = model.coef_
    n_classes = coef.shape[0]
    features = X.columns

    if class_labels is None:
        class_labels = [f"class {i}" for i in range(n_classes)]

    # ì œì™¸í•  í”¼ì²˜ ì²˜ë¦¬
    exclude_features = set(exclude_features or [])

    for i in range(n_classes):
        odds_ratios = np.exp(coef[i])
        coef_df = pd.DataFrame({
            'feature': features,
            'odds_ratio': odds_ratios
        })

        # ì œì™¸í•  feature ì œê±°
        coef_df = coef_df[~coef_df['feature'].isin(exclude_features)]

        # ì˜í–¥ë ¥ ê¸°ì¤€ìœ¼ë¡œ ìƒìœ„ top_n ì„ íƒ
        coef_df['distance_from_1'] = np.abs(coef_df['odds_ratio'] - 1)
        top_coef = coef_df.nlargest(top_n, 'distance_from_1')

        plt.figure(figsize=figsize)
        plt.barh(top_coef['feature'], top_coef['odds_ratio'], 
                 color=np.where(top_coef['odds_ratio'] > 1, 'skyblue', 'salmon'))
        plt.axvline(x=1, color='gray', linestyle='--', label='Odds Ratio = 1')
        plt.xlabel('ì˜¤ì¦ˆë¹„ (Odds Ratio)')
        plt.title(f'Logistic Regression Odds Ratios for {class_labels[i]}')
        plt.gca().invert_yaxis()
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
df.drop(['video_id'], axis=1, inplace=True)
data = df.drop(['view_count_B'], axis=1)   # íƒ€ê²Ÿë³€ìˆ˜ë¥¼ ì œì™¸í•œ ë³€ìˆ˜ë§Œ dataì— ì €ì¥
target = df['view_count_B']                # íƒ€ê²Ÿë³€ìˆ˜ë§Œ targetì— ì €ì¥

# 50:50 ë°ì´í„° ë¶„í• 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.5, random_state=42, stratify=target)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
# Logistic Regression ëª¨í˜• (Default ëª¨ë¸)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(solver='lbfgs',penalty='none',random_state=0,n_jobs=-1)
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Logreg Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Logreg Test set score:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression ëª¨í˜• (Default ëª¨ë¸)
lr = LogisticRegression(solver='lbfgs',penalty='none',random_state=0,n_jobs=-1)

# ê·¸ë¦¬ë“œ ì„œì¹˜
from sklearn.model_selection import GridSearchCV
params = {'solver':['lbfgs', 'saga'], 'penalty':['none']}

grid_lr = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1,)
grid_lr.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_lr.best_score_))
print("GridSearchCV best parameter:", (grid_lr.best_params_))
```

```python
best_clf = grid_lr.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```
