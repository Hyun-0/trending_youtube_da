### ID 중복 Tuple Drop 전 데이터 가공
- (1일 변화량 평균 / 최종 변화량) 비교
- view_count -> view_count_ratio : 높을수록 초기에 조회수가 집중됨 (빠른 확산형 콘텐츠), 낮을수록 시간이 지나며 점진적으로 조회수가 증가 (장기 노출형 콘텐츠)
- likes -> likes_ratio : 높을수록 초기에 강한 호응 (팬 기반, 이슈형 콘텐츠), 낮을수록 좋아요가 점진적으로 증가 (입소문형 콘텐츠)
- dislikes -> dislikes_ratio : 높을수록 초반에 강한 부정 (논란, 부정적 여론의 가능성), 낮을수록 싫어요가 점진적으로 증가 (실망형 콘텐츠, 후속 이슈 발생)
- comment_count -> comment_count_ratio : 높을수록 초기에 댓글이 집중됨 (논란형, 이슈성 콘텐츠), 낮을수록 댓글이 점진적으로 증가 (정보성, 입소문 콘텐츠)

개별 원 데이터 사용 시 값이 쏠린 곳이 많이 존재 -> 정규분포를 따르지 않음
조회수, 좋아요수, 싫어요수 변경값의 평균으로 인급동 재진입 확인

-> 시간이 지날수록 관심이 지속됐는지를 확인하기 위해 평균 사용

```python
# 비율 구하기 위한 변수 1
# 변화량 먼저 계산
df['view_diff'] = df.groupby('video_id')['view_count'].diff()
df['likes_diff'] = df.groupby('video_id')['likes'].diff()
df['dislikes_diff'] = df.groupby('video_id')['dislikes'].diff()
df['comment_count_diff'] = df.groupby('video_id')['comment_count'].diff()

# video_id별 diff 평균 계산 후 각 행에 붙이기
df['view_count_diff_mean'] = df.groupby('video_id')['view_diff'].transform(lambda x: x.dropna().mean())
df['likes_diff_mean'] = df.groupby('video_id')['likes_diff'].transform(lambda x: x.dropna().mean())
df['dislikes_diff_mean'] = df.groupby('video_id')['dislikes_diff'].transform(lambda x: x.dropna().mean())
df['comment_count_diff_mean'] = df.groupby('video_id')['comment_count_diff'].transform(lambda x: x.dropna().mean())
```

```python
# 비율 구하기 위한 변수 2
# groupby + agg로 처음과 마지막 값 차이 계산
agg_df = df.groupby('video_id').agg({
    'view_count': lambda x: x.iloc[-1] - x.iloc[0],
    'likes': lambda x: x.iloc[-1] - x.iloc[0],
    'dislikes': lambda x: x.iloc[-1] - x.iloc[0],
    'comment_count': lambda x: x.iloc[-1] - x.iloc[0]
}).rename(columns={
    'view_count': 'view_count_total_diff',
    'likes': 'likes_total_diff',
    'dislikes': 'dislikes_total_diff',
    'comment_count': 'comment_count_total_diff'
})

# 원본 df에 병합
df = df.merge(agg_df, on='video_id', how='left')
```

## 정규화

```python
cols = [
    'title_length',
    'hours_to_trend',
    'tag_count',
    'view_count_ratio',
    'likes_ratio',
    'dislikes_ratio',
    'comment_count_ratio',
    'description_length'
]

df.drop(columns=cols, inplace=False).to_csv('./youtube_drop_continuous_variables.csv', index=False)

# +-3 체크
# 클수록 좌측, 작을수록 우측으로 치우침
df[cols].skew()
```

```python
# +-10 체크
# 클수록 피크가 높음
df[cols].kurtosis()
```

```python
# 왜도와 첨도 계산 함수
def calculate_skew_kurtosis(df):
    result = []

    for column in df.select_dtypes(include=[np.number]).columns:
        skew_val = df[column].skew()
        kurt_val = df[column].kurtosis()
        result.append({
            'feature': column,
            'skewness': round(skew_val, 3),
            'kurtosis': round(kurt_val, 3)
        })

    return pd.DataFrame(result).sort_values(by=['feature'], ascending=False)


# 왜도와 첨도 값 출력
calculate_skew_kurtosis(df)

# 다양한 변환 적용 함수
def transform_features(df):
    for column in cols:
        if column == 'video_id' or column == 'trending_count':
            continue
        
        skewness = df[column].skew()
        kurt = df[column].kurtosis()
        
        # 절대 왜도와 첨도가 특정 임계값을 넘는 경우 변환 시도
        if abs(skewness) > 3 or abs(kurt) > 10 or abs(skewness) < -3 or abs(kurt) < -10 :
            # print(f'Transforming {column} (skewness={skewness}, kurtosis={kurt})')
            # 로그 변환
            df.loc[:, column + '_log'] = np.log1p(df[column] - df[column].min() + 1)
            # 제곱근 변환
            df.loc[:, column + '_sqrt'] = np.sqrt(df[column] - df[column].min() + 1)
            # 역수 변환
            df.loc[:, column + '_reciprocal'] = 1 / (df[column] - df[column].min() + 1)

    return df


cols = [
    # ID
    'video_id',
    # Target
    'trending_count',
    
    'title_length',
    'hours_to_trend',
    'tag_count',
    'view_count_ratio',
    'likes_ratio',
    'dislikes_ratio',
    'comment_count_ratio',
    'description_length'
]
# 변환 적용
df_skew_kurt = transform_features(df[cols])

# 변환 후 왜도와 첨도 값 출력
df_skew_kurt_printer = calculate_skew_kurtosis(df_skew_kurt)
df_skew_kurt_printer
```

```python
def trim_outliers(df, columns, lower, upper):
    df_copy = df.copy()
    for column in columns:
        df_copy = df_copy[(df_copy[column] >= lower[column]) & (df_copy[column] <= upper[column])]
    return df_copy
  
def trim_outliers(df, column, lower, upper):
    df_copy = df.copy()
    df_copy = df_copy[(df_copy[column] >= lower) & (df_copy[column] <= upper)]
    return df_copy
```

```python
import pandas as pd
import numpy as np

# 평균과 표준편차 계산
mean = trimmed_df['trending_count'].mean()
std = trimmed_df['trending_count'].std()

# 분류 함수 정의
def classify_trending(count):
    if count <= mean - std:
        return 'short'
    elif count >= mean + std:
        return 'long'
    else:
        return 'normal'

# 분류 적용
trimmed_df['trending_group'] = trimmed_df['trending_count'].apply(classify_trending)
trimmed_df.drop(columns=['trending_count'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# 비교할 범주형 변수들
categorical_cols = ['channelId', 'categoryId']
target_col = 'trending_group'

# 카이제곱 통계량과 p-value 저장 리스트
chi2_values = []
p_values = []

for col in categorical_cols:
    contingency = pd.crosstab(df[col], df[target_col])
    chi2, p, dof, expected = chi2_contingency(contingency)
    chi2_values.append(chi2)
    p_values.append(p)

# numpy 배열로 변환 (1행 N열 형태)
chi2_array = np.array([chi2_values])
p_array = np.array([p_values])

# annotation 문자열 만들기 ("χ²: ...\nP: ...")
annot = np.array([[f"χ²: {chi2:.1f}\nP: {p:.4f}" for chi2, p in zip(chi2_values, p_values)]])

# 히트맵 시각화
plt.figure(figsize=(8, 2))
sns.heatmap(
    chi2_array,
    annot=annot,
    fmt='',
    cmap='YlOrRd',
    xticklabels=categorical_cols,
    yticklabels=[target_col],
    cbar=True,
    linewidths=0.5
)
plt.title('Chi-Square Test: Categorical Features vs. trending_group')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



def label_encoding(df, columns):
    df_copy = df.copy()
    class_maps = {}

    for column in columns:
        if pd.api.types.is_integer_dtype(df_copy[column]):
            df_copy[column] = df_copy[column].astype('category')
        else:
            le = LabelEncoder()
            encoded = le.fit_transform(df_copy[column])
            df_copy[column] = pd.Series(encoded, index=df_copy.index).astype('category')
            class_maps[column] = le.classes_

    return df_copy, class_maps




def one_hot_encoding(df, columns, drop_first=False):
    # 항상 복사본으로 작업
    df_copy = df.copy()

    # get_dummies() 실행
    df_copy = pd.get_dummies(df_copy, columns=columns, drop_first=drop_first)

    # True/False → 1/0 변환 (bool → int)
    bool_cols = df_copy.select_dtypes(include=['bool']).columns
    df_copy[bool_cols] = df_copy[bool_cols].astype(int)

    return df_copy
```

## 회귀 모델 지표 해석 가이드
- **MAE (Mean Absolute Error)**  
  - 평균적으로 예측값이 실제값과 얼마나 차이 나는지  
  - 작을수록 좋음

- **MAPE (Mean Absolute Percentage Error)**  
  - 실제값 대비 예측 오차의 평균 비율 (%)  
  - 0~10%: 매우 우수 / 10~20%: 양호 / 20~50%: 보통 / 50% 이상: 나쁨

- **RMSE (Root Mean Squared Error)**  
  - 큰 오차에 더 민감한 평균 오차  
  - 작을수록 좋음  
  - MAE보다 항상 크거나 같음

- **R² (R-squared)**  
  - 0~1 사이 값 (1에 가까울수록 모델이 데이터를 잘 설명함)  
  - 0.7 이상이면 대체로 설명력 우수

- **Adjusted R²**  
  - 설명변수가 많을수록 R²가 과하게 높아지는 걸 보정  
  - R²보다 조금 낮거나 비슷한 수준이어야 정상  
  - **→ 비정상적으로 높거나 낮으면 계산 오류 가능**

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error


def get_mae(model, X_test, y_test):
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    print('{0} MAE: {1}'.format(model.__class__.__name__, np.round(mae, 3)))
    return mae

def get_mape(model, X_test, y_test):
    pred = model.predict(X_test)
    # 예측값이 0에 가까우면 문제가 되므로 작은 수 더함
    mape = np.mean(np.abs((y_test - pred) / (y_test + 1e-8))) * 100
    print('{0} MAPE: {1:.2f}%'.format(model.__class__.__name__, mape))
    return mape

def get_rmse(model, X_test, y_test):
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test , pred)
    rmse = np.sqrt(mse)
    print('{0} 로그 변환된 RMSE: {1}'.format(model.__class__.__name__,np.round(rmse, 3)))
    return rmse

def get_r2(model, X_test, y_test):
    pred = model.predict(X_test)
    r2 = r2_score(y_test , pred)
    print('{0} R²: {1}'.format(model.__class__.__name__,np.round(r2, 3)))
    return r2

def get_adjusted_r2(model, X_test, y_test, n_features):
    pred = model.predict(X_test)
    r2 = r2_score(y_test, pred)
    n = len(y_test)
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - n_features - 1)
    print('{0} Adjusted R²: {1:.3f}'.format(model.__class__.__name__, adj_r2))
    return adj_r2

def get_evaluation_metrics(model, X_test, y_test, n_features):
    mae = get_mae(model, X_test, y_test)
    mape = get_mape(model, X_test, y_test)
    rmse = get_rmse(model, X_test, y_test)
    r2 = get_r2(model, X_test, y_test)
    adj_r2 = get_adjusted_r2(model, X_test, y_test, n_features)
    return mae, mape, rmse, r2, adj_r2
```

```python
def print_evaluation_metrics(mae, mape, rmse, r2, adj_r2):
    print("회귀 모델 평가 결과")
    print("="*40)
    print(f"MAE (Mean Absolute Error): {mae:.3f}")
    print("    → 평균적으로 예측값이 실제값에서 {:.3f} 만큼 차이남".format(mae))
    
    print(f"MAPE (Mean Absolute Percentage Error): {mape:.2f}%")
    if mape < 10:
        quality = "매우 우수한"
    elif mape < 20:
        quality = "양호한"
    elif mape < 50:
        quality = "보통의"
    else:
        quality = "나쁜"
    print(f"    → 예측값의 평균 상대 오차는 {mape:.2f}% 수준  → {quality} 예측 정확도")

    print(f"RMSE (Root Mean Squared Error): {rmse:.3f}")
    print("    → 로그 스케일 기준, 큰 오차에 민감한 평균 오차")

    print(f"R² (R-squared): {r2:.3f}")
    print("    → 모델이 전체 타겟 분산 중 약 {0:.1f}% 를 설명함".format(r2 * 100))
    if r2 >= 0.7:
        print("    → 적절한 설명력")
    elif r2 >= 0.5:
        print("    → 보통 수준의 설명력")
    else:
        print("    → 설명력 낮음")

    print(f"Adjusted R²: {adj_r2:.3f}")
    if adj_r2 > 1 or adj_r2 < 0:
        print("    → 비정상 수치: n_features가 너무 많거나 계산 오류 가능성 있음")
    else:
        print("    → 설명변수 수 보정된 설명력")

    print("="*40)
```

## 분류 모델 지표 해석 가이드
- **Accuracy (정확도)**  
  - 전체 예측 중 정답으로 맞춘 비율  
  - 클래스 분포가 균형잡혀 있을 때 유용  
  - 0.9 이상이면 일반적으로 우수  
  - 불균형 데이터에서는 과대평가될 수 있음

- **Precision (정밀도)**  
  - 양성으로 예측한 것 중 실제로 양성인 비율  
  - False Positive(거짓 양성)를 줄이는 데 효과적  
  - 예: 스팸메일 필터링에서 정상 메일을 스팸으로 잘못 분류하지 않기

- **Recall (재현율)**  
  - 실제 양성 중에서 예측이 맞은 비율  
  - False Negative(거짓 음성)를 줄이는 데 효과적  
  - 예: 암 진단 모델에서 환자를 놓치지 않는 것

- **F1 Score (조화 평균)**  
  - 정밀도와 재현율의 균형을 나타냄  
  - 두 지표가 모두 높을 때 F1 점수도 높음  
  - 0.6 이상: 보통 / 0.7 이상: 양호 / 0.8 이상: 우수

- **ROC AUC (Area Under Curve)**  
  - 모델의 클래스 구분 능력을 나타냄  
  - 0.5 = 무작위 / 1.0 = 완벽한 분리  
  - 0.7~0.8: 보통 / 0.8~0.9: 양호 / 0.9 이상: 매우 우수  
  - 다중 분류에서는 OVO 방식 등으로 확장 가능


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
import numpy as np

def get_accuracy(model, X_test, y_test):
    pred = model.predict(X_test)
    acc = accuracy_score(y_test, pred)
    print('{0} Accuracy: {1:.3f}'.format(model.__class__.__name__, acc))
    return acc

def get_precision(model, X_test, y_test):
    pred = model.predict(X_test)
    prec = precision_score(y_test, pred, average='macro', zero_division=0)
    print('{0} Precision (macro): {1:.3f}'.format(model.__class__.__name__, prec))
    return prec

def get_recall(model, X_test, y_test):
    pred = model.predict(X_test)
    rec = recall_score(y_test, pred, average='macro', zero_division=0)
    print('{0} Recall (macro): {1:.3f}'.format(model.__class__.__name__, rec))
    return rec

def get_f1(model, X_test, y_test):
    pred = model.predict(X_test)
    f1 = f1_score(y_test, pred, average='macro', zero_division=0)
    print('{0} F1 Score (macro): {1:.3f}'.format(model.__class__.__name__, f1))
    return f1

def get_roc_auc(model, X_test, y_test):
    try:
        proba = model.predict_proba(X_test)
        if proba.shape[1] == 2:
            auc = roc_auc_score(y_test, proba[:, 1])
        else:
            auc = roc_auc_score(y_test, proba, multi_class='ovo', average='macro')
        print('{0} ROC AUC: {1:.3f}'.format(model.__class__.__name__, auc))
        return auc
    except:
        print('{0} ROC AUC: Not Available (no predict_proba)'.format(model.__class__.__name__))
        return None

def get_classification_metrics(model, X_test, y_test):
    acc = get_accuracy(model, X_test, y_test)
    prec = get_precision(model, X_test, y_test)
    rec = get_recall(model, X_test, y_test)
    f1 = get_f1(model, X_test, y_test)
    auc = get_roc_auc(model, X_test, y_test)
    return acc, prec, rec, f1, auc
```

```python
def print_classification_metrics(acc, prec, rec, f1, auc):
    print("분류 모델 평가 결과")
    print("="*40)
    print(f"Accuracy (정확도): {acc:.3f}")
    print("    → 전체 예측 중 정답 비율")

    print(f"Precision (정밀도, macro): {prec:.3f}")
    print("    → 예측한 것 중 실제로 맞은 비율 (클래스별 평균)")

    print(f"Recall (재현율, macro): {rec:.3f}")
    print("    → 실제 정답 중 예측이 맞은 비율 (클래스별 평균)")

    print(f"F1 Score (macro): {f1:.3f}")
    print("    → 정밀도와 재현율의 조화 평균")

    if auc is not None:
        print(f"ROC AUC: {auc:.3f}")
        print("    → 클래스 분리 능력 (1에 가까울수록 우수)")
    else:
        print("ROC AUC: 제공 불가 (predict_proba 없음)")

    print("="*40)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_classification_model_comparison_with_values(results_dict, ylim=(0, 1), figsize=(10, 6)):
    """
    분류 모델별 평가 지표를 시각화하고, 각 지표마다 수치를 텍스트 박스로 표시

    Parameters:
        results_dict: {'모델이름': (ACC, PREC, REC, F1, ROC AUC)}
        figsize: 그래프 크기 (기본: (10, 6))
        ylim: y축 범위 (튜플, 기본: (0, 1))
    """
    # 지표 이름
    metric_names = ['ACC', 'PREC', 'REC', 'F1', 'ROC AUC']
    df = pd.DataFrame(results_dict, index=metric_names)

    # 시각화
    plt.figure(figsize=figsize)
    markers = ['o', 's', '^', 'D', 'v']  # 모델 수 많아질 대비

    for i, model_name in enumerate(df.columns):
        plt.plot(df.index, df[model_name], marker=markers[i % len(markers)], label=model_name)

    plt.title('분류 모델별 평가 지표 비교', fontsize=14)
    plt.ylabel('지표값')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()

    # y축 범위 지정
    plt.ylim(*ylim)

    # 각 지표별 박스 텍스트 추가
    for idx, metric in enumerate(df.index):
        y_values = df.loc[metric]
        max_y = y_values.max()
        margin = (ylim[1] - ylim[0]) * 0.05  # 상한-하한의 5%
        text_y = min(max_y + margin, ylim[1] * 1.02)  # ylim 상한을 넘지 않도록 조정

        text = "\n".join([f"{k}: {v:.3f}" for k, v in y_values.items()])
        plt.text(x=idx, y=text_y, s=text,
                 ha='center', va='bottom',
                 fontsize=9,
                 bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="gray", alpha=0.8))

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from util_func import font_setter
font_setter('/System/Library/Fonts/Supplemental/AppleGothic.ttf')

def plot_top_coefficients(model, X, top_n=20, figsize=(10, 6)):
    """
    선형 계열 모델의 계수 중 영향력 큰 변수들을 시각화
    
    Parameters:
        model: 학습된 선형 회귀 모델 (LinearRegression, Ridge, Lasso 등)
        X: 모델 학습에 사용한 DataFrame (더미 포함)
        top_n: 시각화할 상위 변수 개수
        figsize: 그래프 크기
    """
    if not hasattr(model, "coef_"):
        raise ValueError("해당 모델은 coef_ 속성이 없습니다. 선형 모델을 사용하세요.")
    
    coef_df = pd.DataFrame({
        'feature': X.columns,
        'coefficient': model.coef_
    })

    # 절댓값 기준 상위 N개
    top_coef = coef_df.reindex(
        coef_df['coefficient'].abs().sort_values(ascending=False).index
    ).head(top_n)

    # 시각화
    plt.figure(figsize=figsize)
    bars = plt.barh(top_coef['feature'], top_coef['coefficient'], color=np.where(top_coef['coefficient'] > 0, 'skyblue', 'salmon'))
    plt.xlabel('계수값 (Coefficient)')
    plt.title(f'절댓값 기준 Top {top_n} 변수 영향력 (모델: {model.__class__.__name__})')
    plt.gca().invert_yaxis()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
def plot_continuous_coefficients(model, X, figsize=(8, 5)):
    cont_cols = ['title_length', 'hours_to_trend_log', 'tag_count', 'description_length_log', 'PC1']
    
    coef_df = pd.DataFrame({
        'feature': X.columns,
        'coefficient': model.coef_
    })
    cont_df = coef_df[coef_df['feature'].isin(cont_cols)]

    plt.figure(figsize=figsize)
    plt.barh(cont_df['feature'], cont_df['coefficient'], color=np.where(cont_df['coefficient'] > 0, 'skyblue', 'salmon'))
    plt.xlabel('계수값 (Coefficient)')
    plt.title('연속형 변수 영향력')
    plt.gca().invert_yaxis()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
def plot_categorical_coefficients(model, X, prefix, top_n=None, figsize=(10, 6)):
    """
    특정 범주형 변수(prefix)의 더미 변수 계수를 시각화
    
    Parameters:
        model: 학습된 선형 회귀 모델
        X: 학습에 사용된 DataFrame (더미 포함)
        prefix: 변수 prefix (e.g., 'channelId', 'categoryId')
        top_n: 상위 n개만 시각화 (None이면 전부 시각화)
        figsize: 그래프 크기
    """
    coef_df = pd.DataFrame({
        'feature': X.columns,
        'coefficient': model.coef_
    })

    # 해당 prefix로 시작하는 컬럼 필터링
    cat_df = coef_df[coef_df['feature'].str.startswith(prefix)]

    # top_n 적용
    if top_n is not None:
        cat_df = cat_df.reindex(
            cat_df['coefficient'].abs().sort_values(ascending=False).index
        ).head(top_n)

    plt.figure(figsize=figsize)
    plt.barh(cat_df['feature'], cat_df['coefficient'], color=np.where(cat_df['coefficient'] > 0, 'skyblue', 'salmon'))
    plt.xlabel('계수값 (Coefficient)')
    plt.title(f"{prefix} 변수 영향력" + (f" (Top {top_n})" if top_n else ""))
    plt.gca().invert_yaxis()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
    
def plot_categorical_coefficients_clean(model, X, prefix, figsize=(12, 6)):
    """
    prefix로 시작하는 더미 변수들 계수를 '클린' 스타일로 시각화
    - label 제거
    - grid 제거
    - 선명한 색상만 강조

    Parameters:
        model: 학습된 선형 모델
        X: DataFrame with feature names
        prefix: 범주형 변수 prefix (e.g., 'channelId', 'categoryId')
        figsize: 시각화 크기
    """

    coef_df = pd.DataFrame({
        'feature': X.columns,
        'coefficient': model.coef_
    })

    cat_df = coef_df[coef_df['feature'].str.startswith(prefix)]

    plt.figure(figsize=figsize)
    plt.bar(range(len(cat_df)), cat_df['coefficient'], 
            color=np.where(cat_df['coefficient'] > 0, 'skyblue', 'salmon'))

    plt.title(f"{prefix} 변수 영향력 (전체)", fontsize=14)
    plt.xlabel('채널 순번', fontsize=12)
    plt.ylabel('계수값 (Coefficient)', fontsize=12)

    # 눈금, 라벨 제거로 깔끔하게
    plt.xticks([])  # X축 눈금 제거
    plt.yticks(fontsize=10)
    plt.box(False)  # 그래프 테두리 제거
    plt.tight_layout()
    plt.show()
```

```python
def show_plot(model, X_test):
    plot_top_coefficients(model, X_test, top_n=20, figsize=(7, 4))
    plot_continuous_coefficients(model, X_test, figsize=(7, 4))
    # plot_categorical_coefficients(lr_reg, X_test, 'channelId', top_n=None, figsize=(10, 6))
    plot_categorical_coefficients_clean(model, X_test, 'channelId', figsize=(7, 4))
    plot_categorical_coefficients(model, X_test, 'categoryId', top_n=None, figsize=(7, 4))
```

```python
def plot_logistic_coefficients(model, X, class_labels=None, figsize=(10, 6), top_n=20):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    if not hasattr(model, "coef_"):
        raise ValueError("해당 모델은 coef_ 속성이 없습니다.")
    
    coef = model.coef_  # shape: (n_classes, n_features)
    n_classes = coef.shape[0]
    features = X.columns

    class_labels = class_labels or [f"class {i}" for i in range(n_classes)]

    for i in range(n_classes):
        coef_df = pd.DataFrame({
            'feature': features,
            'coefficient': coef[i]
        })
        top_coef = coef_df.reindex(
            coef_df['coefficient'].abs().sort_values(ascending=False).index
        ).head(top_n)

        plt.figure(figsize=figsize)
        plt.barh(top_coef['feature'], top_coef['coefficient'], 
                 color=np.where(top_coef['coefficient'] > 0, 'skyblue', 'salmon'))
        plt.xlabel('계수값')
        plt.title(f'Logistic Regression Coefficients for {class_labels[i]}')
        plt.gca().invert_yaxis()
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
def plot_logistic_odds_ratios(model, X, class_labels=None, figsize=(10, 6), top_n=20, exclude_features=None):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    if not hasattr(model, "coef_"):
        raise ValueError("해당 모델은 coef_ 속성이 없습니다.")
    
    coef = model.coef_
    n_classes = coef.shape[0]
    features = X.columns

    if class_labels is None:
        class_labels = [f"class {i}" for i in range(n_classes)]

    # 제외할 피처 처리
    exclude_features = set(exclude_features or [])

    for i in range(n_classes):
        odds_ratios = np.exp(coef[i])
        coef_df = pd.DataFrame({
            'feature': features,
            'odds_ratio': odds_ratios
        })

        # 제외할 feature 제거
        coef_df = coef_df[~coef_df['feature'].isin(exclude_features)]

        # 영향력 기준으로 상위 top_n 선택
        coef_df['distance_from_1'] = np.abs(coef_df['odds_ratio'] - 1)
        top_coef = coef_df.nlargest(top_n, 'distance_from_1')

        plt.figure(figsize=figsize)
        plt.barh(top_coef['feature'], top_coef['odds_ratio'], 
                 color=np.where(top_coef['odds_ratio'] > 1, 'skyblue', 'salmon'))
        plt.axvline(x=1, color='gray', linestyle='--', label='Odds Ratio = 1')
        plt.xlabel('오즈비 (Odds Ratio)')
        plt.title(f'Logistic Regression Odds Ratios for {class_labels[i]}')
        plt.gca().invert_yaxis()
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd       # pandas library 불러내기
import numpy as np        # numpy library 불러내기

#pd.set_option('display.max_columns', None)   # 결과물로 보여주는 열 갯수 최대화
#pd.set_option('display.max_rows', None)      # 결과물로 보여주는 행 갯수 최대화

# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project/KR_youtube_trending_data_4.csv', lineterminator='\n')
df = pd.read_csv('./KR_youtube_trending_data_4.csv', lineterminator='\n')
                                             # 데이터파일을 불러와서 데이터프레임(df)에 저장
df.head(3)                                   # 데이터프레임(df)를 3행까기 보여주기
```

```python
df.drop(['video_id'], axis=1, inplace=True)
data = df.drop(['view_count_B'], axis=1)   # 타겟변수를 제외한 변수만 data에 저장
target = df['view_count_B']                # 타겟변수만 target에 저장

# 50:50 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.5, random_state=42, stratify=target)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
# Logistic Regression 모형 (Default 모델)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(solver='lbfgs',penalty='none',random_state=0,n_jobs=-1)
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Logreg Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Logreg Test set score:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 모형 (Default 모델)
lr = LogisticRegression(solver='lbfgs',penalty='none',random_state=0,n_jobs=-1)

# 그리드 서치
from sklearn.model_selection import GridSearchCV
params = {'solver':['lbfgs', 'saga'], 'penalty':['none']}

grid_lr = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1,)
grid_lr.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_lr.best_score_))
print("GridSearchCV best parameter:", (grid_lr.best_params_))
```

```python
best_clf = grid_lr.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```
